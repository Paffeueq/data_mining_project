> **Zadanie 1**  Znajdź:
* państwa o zbliżonej liczbie ludności do Polski
* najbardziej podobne do Polski państwa świata pod względem zarówno powierzchni jak i liczby ludności

```python


COL_POP = panstwa.columns[2] # population column name
COL_AREA = panstwa.columns[3] # area column name

# Values for Poland
polska_row = panstwa.loc[panstwa['Country'] == 'Poland'].iloc[0]
pop_pl   = polska_row.iloc[2]
area_pl  = polska_row.iloc[3]

# 1a

# Setting population range ±10%
pop_min = pop_pl * 0.9
pop_max = pop_pl * 1.1

podobna_populacja = panstwa.loc[
    (panstwa.iloc[:, 2].between(pop_min, pop_max, inclusive='both')) &
    (panstwa['Country'] != 'Poland'),
    ['Country', COL_POP]
]

print("Państwa o zbliżonej populacji względem Polski:\n")
display(podobna_populacja)

# 1b
tmp = panstwa.loc[panstwa['Country'] != 'Poland', ['Country', COL_POP, COL_AREA]].copy()

# Calculate similarity score based on population and area differences
tmp['PopDiff']  = (tmp[COL_POP]  - pop_pl).abs()
tmp['AreaDiff'] = (tmp[COL_AREA] - area_pl).abs()

# Combine differences into a single similarity score
tmp['SimilarityScore'] = tmp['PopDiff'] + tmp['AreaDiff']

# Sort by similarity score and select top 8 most similar countries
most_similar = tmp.sort_values('SimilarityScore', ascending=True).head(8)

print("\nNajbardziej podobne państwa do Polski (populacja + powierzchnia) TOP 8:\n")
display(most_similar[['Country', COL_POP, COL_AREA, 'SimilarityScore']])
```

```python
# === ZBIÓR D =========================


print("Opis dla zbioru d:")
display(d.describe())

# Histogramy
d.hist(figsize=(8, 4), bins=10, color="lightblue", edgecolor="black")
plt.suptitle("Histogramy dla pełnego zbioru d")
plt.show()

# Wykresy pudełkowe (boxplot)
plt.figure(figsize=(6, 4))
sns.boxplot(data=d.iloc[:, :2])
plt.title("Wykres pudełkowy – pełny zbiór d")
plt.show()

# Wykres punktowy (pairplot = scatter matrix)
sns.pairplot(d, hue="klasa", diag_kind="hist", palette="Set2")
plt.suptitle("Macierz wykresów punktowych (pairplot) – zbiór d", y=1.02)
plt.show()

# --- Dane dla poszczególnych klas ---
klasy = d["klasa"].unique()
for k in klasy:
    print(f"\n Miary dla {k}:")
    display(d[d["klasa"] == k].describe())

    # Histogramy dla klasy
    d[d["klasa"] == k].iloc[:, :2].hist(figsize=(8, 4), bins=10, color="lightgreen", edgecolor="black")
    plt.suptitle(f"Histogramy – {k}")
    plt.show()

    # Boxplot dla klasy
    plt.figure(figsize=(6, 4))
    sns.boxplot(data=d[d["klasa"] == k].iloc[:, :2])
    plt.title(f"Wykres pudełkowy – {k}")
    plt.show()

    # Wykres punktowy dla klasy
    plt.figure(figsize=(6, 4))
    sns.scatterplot(
        data=d[d["klasa"] == k],
        x="atrybut1", y="atrybut2",
        hue="klasa", palette="tab10", s=60
    )
    plt.title(f"Wykres punktowy – {k}")
    plt.legend()
    plt.show()


# === ZBIÓR IRIS ======================

print("\n\n Miary dla zbioru iris:")
display(iris.describe())

# Histogramy
iris.hist(figsize=(8, 6), bins=10, color="lightblue", edgecolor="black")
plt.suptitle("Histogramy dla pełnego zbioru iris")
plt.show()

# Boxplot
plt.figure(figsize=(6, 4))
sns.boxplot(data=iris.iloc[:, :4])
plt.title("Wykres pudełkowy – pełny zbiór iris")
plt.show()

# Pairplot (scatter matrix)
sns.pairplot(iris, hue="species", diag_kind="hist", palette="Set2")
plt.suptitle("Macierz wykresów punktowych (pairplot) – zbiór iris", y=1.02)
plt.show()

# --- Dane dla poszczególnych gatunków ---
gatunki = iris["species"].unique()
for g in gatunki:
    print(f"\n Miary dla gatunku: {g}")
    display(iris[iris["species"] == g].describe())

    # Histogramy dla gatunku
    iris[iris["species"] == g].iloc[:, :4].hist(figsize=(8, 6), bins=10, color="lightgreen", edgecolor="black")
    plt.suptitle(f"Histogramy – {g}")
    plt.show()

    # Boxplot dla gatunku
    plt.figure(figsize=(6, 4))
    sns.boxplot(data=iris[iris["species"] == g].iloc[:, :4])
    plt.title(f"Wykres pudełkowy – {g}")
    plt.show()

    # Wykres punktowy dla gatunku
    plt.figure(figsize=(6, 4))
    sns.scatterplot(
        data=iris[iris["species"] == g],
        x="sepal_length", y="sepal_width",
        hue="species", palette="tab10", s=60
    )
    plt.title(f"Wykres punktowy – {g}")
    plt.legend()
    plt.show()
```

Histogramy pokazują rozkład wartości i wartości skrajne, a boxploty umożliwiają szybkie odczytanie mediany, rozstępu i obserwacji odstających. Pairploty ujawniają zależności między cechami oraz stopień separacji klas lub gatunków.

```python

# zbior d

plt.figure(figsize=(6, 4))
sns.pairplot(data=d, vars=['atrybut1', 'atrybut2'], hue='klasa', palette='Set2', diag_kind='hist')
plt.suptitle('Zbiór d – relacje między atrybutami', y=1.03)
plt.show()

# Obliczenie i wizualizacja korelacji
corr_d = d[['atrybut1', 'atrybut2']].corr(method='pearson')
print("🔹 Korelacje między atrybutami w zbiorze d:\n")
print(corr_d.round(3))

plt.figure(figsize=(4, 4))
sns.heatmap(corr_d, annot=True, cmap='RdYlBu', fmt='.2f', square=True)
plt.title('Zbiór d – macierz korelacji')
plt.tight_layout()
plt.show()



 # zbior iris

iris_features = iris.select_dtypes(include='number')

sns.pairplot(data=iris, vars=iris_features.columns, hue='species', palette='husl', diag_kind='hist')
plt.suptitle('Zbiór iris – relacje między cechami', y=1.03)
plt.show()

corr_iris = iris_features.corr(method='pearson')
print("\n🔹 Korelacje między cechami w zbiorze iris:\n")
print(corr_iris.round(3))

plt.figure(figsize=(6, 5))
sns.heatmap(corr_iris, annot=True, cmap='RdYlBu', fmt='.2f', square=True)
plt.title('Zbiór iris – macierz korelacji')
plt.tight_layout()
plt.show()
```

Korelacja mówi, jak mocno dwie cechy zmieniają się razem.
Jeśli punkty leżą wzdłuż prostej – korelacja jest silna (|ρ|≈1), a jeśli są rozrzucone – korelacji prawie nie ma (|ρ|≈0).

W zbiorze d punkty tworzą oddzielne grupy (klastry), ale nie układają się w linię, dlatego korelacja między atrybut1 i atrybut2 jest prawie zerowa (ρ≈−0.03).

W zbiorze iris widać silną dodatnią zależność między długością i szerokością płatka (petal_length, petal_width, ρ≈0.96) – im dłuższy płatek, tym szerszy.
